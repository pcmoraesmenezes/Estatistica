```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
```

```python
# .ods 
df = pd.read_excel('trab.xls')
```

```python
x1 = df['X1']
y1 = df['Y1']
x2 = df['X2']
y2 = df['Y2']
```

```python
x1_mean = round(x1[1:].mean(), 3)
x1_std = round(x1[1:].std(), 3)
coeficiente_variacao_x1 = round(x1_std / x1_mean, 3)
x1_median = round(x1[1:].median(), 3)

print('X1:\n Mean: {:.3f}\n Std: {:.3f}\n Coeficiente de variação: {:.3f}\n Median: {:.3f}'.format(x1_mean, x1_std, coeficiente_variacao_x1, x1_median))
```

```python
y1 = pd.to_numeric(y1, errors='coerce')
y1_mean = round(y1[1:].mean(), 3)
y1_std = round(y1[1:].std(), 3)
y1_median = round(y1[1:].median(), 3)
coeficiente_variacao_y1 =round (y1_std / y1_mean, 3)

print('Y1:\n Mean: {}\n Std: {}\n Coeficiente de variação: {}\n Mediana: {} '.format(y1_mean, y1_std, coeficiente_variacao_y1, y1_median))
```

```python
x2 = pd.to_numeric(x2, errors='coerce')
x2_mean = round(x2[1:].mean(), 3)
x2_std = round(x2[1:].std(), 3)
x2_median = round(x2[1:].median(), 3)
coeficiente_variacao_x2 = round(x2_std / x2_mean, 3)

print('X2:\n Mean: {}\n Std: {}\n Coeficiente de variação: {}\n Mediana: {}'.format(x2_mean, x2_std, coeficiente_variacao_x2, x2_median))
```

```python
y2 = pd.to_numeric(y2, errors='coerce')
y2_mean = round(y2[1:].mean(), 3)
y2_std = round(y2[1:].std(), 3)
y2_median = round(y2[1:].median(), 3)
coeficiente_variacao_y2 = round(y2_std / y2_mean, 3)

print('Y2:\n Mean: {}\n Std: {}\n Coeficiente de variação: {}\n Mediana: {}'.format(y2_mean, y2_std, coeficiente_variacao_y2, y2_median))
```

```python
x1 = pd.to_numeric(x1, errors='coerce')

x1_coeficiente_de_correlacao = round(x1[1:].corr(y1[1:]), 3)

print('Coeficiente de correlação entre X1 e Y1: {}'.format(x1_coeficiente_de_correlacao))
```

```python
alpha = 0.01

coef_corr, p_valor = stats.pearsonr(x1[1:], y1[1:])

print(f'Coeficiente de Correlação: {coef_corr:.3f}')
print(f'P-valor: {p_valor:.3f}')

if p_valor < alpha:
    print('Rejeitar a hipótese nula. Há evidências de correlação entre x1 e y1.')
else:
    print('Não rejeitar a hipótese nula. Não há evidências suficientes para afirmar correlação entre x1 e y1.')
```

```python
x2_coeficiente_de_correlacao = round(x2[1:].corr(y2[1:]), 3)

print('Coeficiente de correlação entre X2 e Y2: {}'.format(x2_coeficiente_de_correlacao))
```

```python
alpha = 0.01

coef_corr, p_valor = stats.pearsonr(x2[1:], y2[1:])

print(f'Coeficiente de Correlação: {coef_corr:.3f}')
print(f'P-valor: {p_valor:.3f}')

if p_valor < alpha:
    print('Rejeitar a hipótese nula. Há evidências de correlação entre x2 e y2.')
else:
    print('Não rejeitar a hipótese nula. Não há evidências suficientes para afirmar correlação entre x2 e y2.')
```

```python
n = len(x1[1:])
sum_x1 = x1[1:].sum()
sum_y1 = y1[1:].sum()

sum_xy = np.sum(x1*y1)
sum_x_squared = np.sum(x1**2)

m = round((n*sum_xy - sum_x1*sum_y1) / (n*sum_x_squared - sum_x1**2), 3)
b = (sum_y1 - m*sum_x1) / n

print('Coeficiente angular: {}'.format(m))
```

```python
plt.figure(figsize=(5,5), dpi=100)
plt.title('Coeficiente entre X1 e Y1')
plt.scatter(x1[1:], y1[1:], marker='o', color='blue')
plt.xlabel('X1')
plt.ylabel('Y1')
plt.plot(x1[1:], m*x1[1:] + b, color='red')
```

```python
n = len(x2[1:])
sum_x2 = x2[1:].sum()
sum_y2 = y2[1:].sum()

sum_xy = np.sum(x2*y2)
sum_x_squared = np.sum(x2**2)

m = round((n*sum_xy - sum_x2*sum_y2) / (n*sum_x_squared - sum_x2**2), 3)
b = (sum_y2 - m*sum_x2) / n

print('Coeficiente angular: {}'.format(m))
```

```python
plt.figure(figsize=(5,5), dpi=100)
plt.title('Coeficiente entre X2 e Y2')
plt.scatter(x2[1:], y2[1:], marker='o', color='blue')
plt.xlabel('x2')
plt.ylabel('y2')
plt.plot(x2[1:], m*x2[1:] + b, color='red')
```

```python
from scipy.stats import linregress

slope, intercept, r_value, p_value, std_err = linregress(x1[1:], y1[1:])

r_squared = round(r_value**2,3)

print('Coeficiente de determinação: {}'.format(r_squared))
```

```python
slope, intercept, r_value, p_value, std_err = linregress(x2[1:], y2[1:])

r_squared = round(r_value**2,3)

print('Coeficiente de determinação: {}'.format(r_squared))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
modelo = LinearRegression()
X = np.column_stack((x1[1:], x2[1:]))

modelo.fit(X, y1[1:])

Q3_X1 = np.percentile(x1[1:], 75)
Q3_X2 = np.percentile(x2[1:], 75)

previsao = modelo.predict([[Q3_X1, Q3_X2]])

print(f'A previsão da quantidade total de esforço para Q3_X1 e Q3_X2 é: {previsao[0]:.3f}')
```
